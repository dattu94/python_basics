## Filtering columns with loc

Use iloc for integer-positional selection (always reliable for positions).

Use loc for label-based selection (reliable for actual index labels).

df.loc[0:3,] only works as expected if your DataFrame's index contains the labels 0, 1, 2, 3. 
If your index is strings or custom non-sequential numbers, it will likely raise a KeyError.

```python
columns = ['Employee_ID','Age', 'Location']
df_filter_col = df.loc[0:3,columns]
df_filter_col
```

```python
df_filter_col2 = df.iloc[:,:-4]
df_filter_col2
```

## if the comma seperated are count is differnt then follow this approach

```python
# Example df (replace with your actual df if it's different):
data = {
    'Name': ['sree Dattu', 'harish', 'vigrat', 'Surya', 'Priya', 'Kiran', 'Rahul'],
    'Address': [
        'Chanda nagar, hyderabad',
        'moosapet, hyderabad',
        'Javvadhivari street,penugonda',
        'Jvr colony, palakol',
        '123 Main St, Anytown, CA, 90210',
        'P.O. Box 123, Some City',
        'Rural Route 1'
    ]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
split_Address_parts = df['Address'].str.split(',',expand = True)
```

```python
split_Address_parts
```

```python
split_Address_parts = split_Address_parts.apply(lambda x: x.str.strip() if x.dtype == 'Object' else x) # this willl strip the empty spaces
```

```python
num_parts = split_Address_parts.shape[1]
```

```python
new_colums = [f'Address_part{i+1}' for i in range(num_parts)]
```

```python
split_Address_parts.columns = new_colums
```

```python
final_df = pd.concat([df, split_Address_parts],axis =1)
```
